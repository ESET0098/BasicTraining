```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
house  = pd.read_csv(r"C:\Users\PC\Downloads\Salary_dataset.csv")
```

```python
house.head()
```

```python
house = house.drop(['Unnamed: 0'],axis=1)
```

```python
house
```

```python
y = house['Salary']
x = house[['YearsExperience']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.8,random_state=22)
```

```python
x_train = x_train.values.reshape(-1,1)
x_test = x_test.values.reshape(-1,1)
y_train = y_train.values.reshape(-1,1)
y_test = y_test.values.reshape(-1,1)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(y_test,y_pred,color='red')
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.title("Actual Salary vs Predicted Salary")
plt.show()
```

```python
from sklearn.metrics import r2_score as r2
r2score = r2(y_test,y_pred)
print("Accuracy is :",r2score)
print("Mean squred Error:",mean_squared_error(y_test,y_pred))
```
